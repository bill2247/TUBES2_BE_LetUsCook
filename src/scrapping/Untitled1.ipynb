```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://little-alchemy.fandom.com/wiki/Elements_(Little_Alchemy_2)'
page = requests.get(url)
soup = BeautifulSoup(page.text,  'html')
```

```python
elements = soup.find_all('tr')
elm =[]
pictures = []
for i in range (len(elements)):
  if (i < 3):
    continue
  element = elements[i].find_all('td')
  if (len(element) < 2):
    continue
  element = element[0].find_all('a')
  if (len(element) < 2):
    continue
  picture = element[0].get('href')
  if (picture == None):
    continue
  element = element[1].get('title')
  if (element == None):
    continue
  print(element)
  print(picture)
  elm.append(element)
  pictures.append(picture)
print(len(elm))
print(len(pictures))
```

```python
mapperElmIdx = {}
mapperPctIdx = {}
mapperIdxElm = {}
mapperIdxPct = {}

for i in range (len(elm)):
  mapperElmIdx[i] = elm[i]
  mapperPctIdx[i] = pictures[i]
  mapperIdxElm[elm[i]] = i
  mapperIdxPct[pictures[i]] = i
i = len(elm)
print(mapperElmIdx)
print(mapperPctIdx)
print(mapperIdxElm)
print(mapperIdxPct)
print(len(mapperElmIdx))
print(len(mapperPctIdx))
print(len(mapperIdxElm))
print(len(mapperIdxPct))
```

```python
rows, cols = 720,720
matrix = [[-1 for _ in range(cols)] for _ in range(rows)]
mapRecipe1 = {}
mapRecipe2 = {}

newElmts = soup.find_all('tr')
count = 1
coutrecipe = 1
for i in range (len(newElmts)):
  if (i < 3):
    continue
  elmt = newElmts[i].find_all('td')
  if (len(elmt) < 2):
    continue
  result = elmt[0].find_all('a')
  if (len(result) < 1):
    continue
  result = result[1].get('title')
  # mencari index dari result pada mapperElmIdx
  idx = mapperIdxElm[result]
  print(idx, result, " = ", mapperElmIdx[idx], " = ", idx)
  elmt = elmt[1].find_all('li')
  if (len(elmt) < 1):
    continue
  for j in range (len(elmt)):
    print('    ', elmt[j].get_text())
    recipe = elmt[j].get_text()
    # recipe = recipeA + " + " + recipeB
    recipe = recipe.split('+')
    recipeA = recipe[0].strip()
    recipeB = recipe[1].strip()
    if not (recipeA in mapperIdxElm and recipeB in mapperIdxElm):
      print("     fail")
      continue
    idxA = mapperIdxElm[recipeA]
    idxB = mapperIdxElm[recipeB]
    print("     ", idxA, " : ", recipeA)
    print("     ", idxB, " : ", recipeB)
    coutrecipe += 1
    matrix[idxB][idxA] = idx
    matrix[idxA][idxB] = idx
    if idx not in mapRecipe1:
      mapRecipe1[idx] = []  # buat list kosong kalau belum ada

    mapRecipe1[idx].append((idxA, idxB))  # pakai append, bukan add
    mapRecipe2[(idxA,idxB)] = idx
    mapRecipe2[(idxB,idxA)] = idx
  count += 1
print(count)
print(coutrecipe)
print(matrix)
print(mapRecipe1)
print(mapRecipe2)
```
